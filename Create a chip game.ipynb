```python
#Create a chip game

import random
import time
import sys

print('歡迎光臨，請創建帳號。')
name = input('用戶名：')
pd = input('密碼：')
pd2 = input('請再次輸入密碼：')

while(pd != pd2):
    pd2 = input('輸入錯誤，請再試一次')
time.sleep(1)
print('帳號創建成功，請重新登入')
in1 = input('用戶名')
in2 = input('密碼')

while(in1!= name or in2 != pd):
    print('輸入錯誤，請再次一次')
    in1 = input('用戶名')
    in2 = input('密碼')
time.sleep(1)
print(name,' 歡迎回來！\n\n')
bargaining_chip = 5000
time.sleep(1)
while(1):
    print('請選擇遊戲種類：\nA 比大小\nB 猜數字（共三次機會）\nC 圈叉\n')
    print('或者輸入Q退出遊戲')
    print('-'*10)
    print('目前的籌碼為：',bargaining_chip)
    in3 = input('請選擇遊戲')
    
    if bargaining_chip < 500:#籌碼不足 需儲值
        print('籌碼不足')
        print('-'*10)
        in4 = input('請問是否儲值？（輸入Y或者Ｎ）')
        if in4 == 'Y' or in4 == 'y':
            in5 = int(input('請輸入儲值金額：'))
            print('儲值成功！')
            bargaining_chip += in5
            print('目前籌碼為：',bargaining_chip, '\n')
        elif in4 == 'N' or in4 == 'n':
            print('謝謝光臨 ', name)
            time.sleep(3)
            break
           
    def Than_size(): #比大小
        min, max = 1, 100 
        win_control = int(input('請輸入莊家贏的比率(0-100)之間：')) #可調整數值比例，以控制玩家勝率，此為莊家輸入。
        bet_a = int(input('請輸入賭金(最少為500)：'))
        while(bet_a < 500 or bet_a > bargaining_chip):
            bet_a = int(input('請輸入賭金(最少為500)：'))
        while(1):
            player_ans = input('猜大小遊戲：輸入L表示大，輸入S表示小，若要退出則請輸入Q')
            if player_ans == 'Q': 
                print('退出')
                break
            than_num = random.randint(min, max)
            if than_num > win_control:
                print('恭喜答對了！獲得獎金：', bet_a)
                print('目前籌碼為：', bargaining_chip + bet_a, '\n')
                time.sleep(3)
                return bargaining_chip + bet_a
                break
            else :
                print('挑戰失敗')
                print('目前籌碼為：',bargaining_chip - bet_a, '\n')
                time.sleep(3)
                return bargaining_chip - bet_a
                break
        
    if in3 == 'A' or in3 == 'a':
        bargaining_chip = Than_size()
        
    
    def Guess_number(): #猜數字
        ans_num = random.randint(1,100)
        ant = 0
        bet_b = int(input('請輸入賭金(最少為500)：'))
        while(bet_b < 500 or bet_b > bargaining_chip):
            bet_b = int(input('請輸入賭金(最少為500)：'))
        while(1):
            player_num = int(input('請輸入數字 1 - 100 : '))
            if player_num > ans_num:
                print('數字太大了')
                ant += 1
            if player_num < ans_num:
                print('數字太小了')
                ant += 1
            if ant == 3:
                print('挑戰失敗')
                print('目前籌碼為：',bargaining_chip - bet_b, '\n')
                time.sleep(3)
                return bargaining_chip - bet_b
                break
            if player_num == ans_num:
                print('恭喜答對了！獲得獎金：', bet_b)
                print('目前籌碼為：', bargaining_chip + bet_b, '\n')
                time.sleep(3)
                return bargaining_chip + bet_b
                break
          
    if in3 == 'B' or in3 == 'b':
        bargaining_chip = Guess_number()
     
    if in3 == 'Q': #跳出
        print('謝謝光臨 ', name)
        time.sleep(3)
        break
    
    class TicTacToe():#圈叉遊戲
        def __init__(self, height, width):
            self.width = width
            self.height = height
            self.board = [[' ']*self.height for i in range(self.width)]
    # 初始化棋盤
        def iniBoard(self):
            for i in range(self.width):
                for j in range(self.height):
                    self.board[i][j]=' '
    #棋盤的圖形
        def drawBoard(self) -> None:
            HLINE =  ' ' * 3 + '+---' * self.width  + '+'
            VLINE = (' ' * 3 +'|') *  (self.width +1)
            title = '     1'
            for i in range(1,self.width):
                title += ' ' * 3 +str(i+1)
            print(title)
            print(HLINE)
            for y in range(self.height):
                print(VLINE)
                print(y+1, end='  ')
                for x in range(self.width):
                    print(f'| {self.board[x][y]}', end=' ')
                print('|')
                print(VLINE)
                print(HLINE)
    
        def isOnBoard(self, x, y):
            return 0 <= x < self.width and 0 <= y < self.height
    #檢查是否為合法棋步
        def isValidMove(self, tile, x, y):
            return self.isOnBoard(x, y) and self.board[x][y]==' '
    # 回傳現在走的所有合法棋步
        def getValidMoves(self, tile):
            return [[x, y] for x in range(self.width) for y in range(self.height) if self.isValidMove(tile, x, y)]
    #ai下棋   
    class TicTacToeAI(TicTacToe):
        def __init__(self, board, height, width):
            super().__init__(height, width)
            self.board = board   
    # 給定盤面board，回傳電腦的選擇
        def getComputerMove(self, computerTile):
            possibleMoves = self.getValidMoves(computerTile)
            random.shuffle(possibleMoves) #給個隨機值
            return possibleMoves[0]


    class Game(TicTacToe):
        def __init__(self, height, width):
            super().__init__(height, width)
            self.turn = 'player'
            self.ai = TicTacToeAI(self.board,self.height, self.width)
    #玩家輸入棋步
        def getPlayerMove(self, playerTile):
            DIGITS = [str(i) for i in range(1,10)]
            while True:
                move = input('輸入棋步').lower()
                if move in {'quit'}:
                    return move
                if len(move) == 2 and move[0] in DIGITS and move[1] in DIGITS:
                    x = int(move[0]) - 1
                    y = int(move[1]) - 1
                    if self.isValidMove(playerTile, x, y):
                        break
                print('重複，換一個')
            return [x, y]
    # 判斷有沒有人贏了(運用剪枝法，網路上查的我也不是很懂，但是能用...)
        def check_TicTacToe(self):
            rows = list(map(''.join,self.board))
            cols = list(map(''.join, zip(*rows)))
            diags = list(map(''.join, zip(*[(r[i], r[2 - i]) for i, r in enumerate(rows)])))
            lines = rows + cols + diags

            if 'XXX' in lines:
                return 'X'  
            if 'OOO' in lines:
                return 'O' 
            return 'D'


        def gameloop(self):
            while True:
                self.iniBoard()
                playerTile, computerTile = ['X', 'O']
                bet_c = int(input('請輸入賭金(最少為500)：'))
                while(bet_c < 500 or bet_c > bargaining_chip):
                    bet_c = int(input('請輸入賭金(最少為500)：'))
                print('玩家先手' if self.turn == 'player' else '電腦先手')

                while True:
                    playerValidMoves = self.getValidMoves(playerTile)
                    computerValidMoves = self.getValidMoves(computerTile)
        #沒有棋可以下的時候就跳出
                    if not playerValidMoves and not computerValidMoves or self.check_TicTacToe()!='D':
                        break

                    if self.turn == 'player' and playerValidMoves:
                        self.drawBoard()
                        move = self.getPlayerMove(playerTile)
                        self.board[move[0]][move[1]] = playerTile
                    elif self.turn == 'computer' and computerValidMoves:
                        self.drawBoard()
                        input('按enter輪到電腦')
                        x, y = self.ai.getComputerMove(computerTile)
                        self.board[x][y] = computerTile
                    self.turn = 'player' if self.turn=='computer' else 'computer'
        #顯示結果
                self.drawBoard()
                result = self.check_TicTacToe()
                if result=='X':
                    print('恭喜答對了！獲得獎金：', bet_c)
                    print('目前籌碼為：', bargaining_chip + bet_c, '\n')
                    time.sleep(3)
                    return bargaining_chip + bet_c
                    break
                elif result=='O':
                    print('挑戰失敗')
                    print('目前籌碼為：',bargaining_chip - bet_c, '\n')
                    time.sleep(3)
                    return bargaining_chip - bet_c
                    break
                else:
                    print('平手')

                if not self.playAgain():
                    break

    if in3 == 'C' or in3 == 'c':
        reversi = Game(3,3)
        bargaining_chip = reversi.gameloop()
        
     
    if in3 == 'Q' or in3 == 'q': #跳出
        print('謝謝光臨 ', name)
        time.sleep(3)
        break
```
